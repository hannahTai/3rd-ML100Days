```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
row_counter, column_counter = app_train.shape
print('row數為', row_counter)
print('column數為', column_counter)
```

#### 列出所有欄位

```python
columns = app_train.columns
for i in range(len(columns)): 
    print('Column {:3d} : {}'.format(i+1, columns[i])) 
```

#### 截取部分資料

```python
## df.iloc[row_begin:row_end, column_begin:column_end]
app_train.iloc[:5, :10] 
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

#### 指定特定欄位, 對欄位資料作排序, 選取前10筆

```python
app_train[['SK_ID_CURR', 'AMT_INCOME_TOTAL', 'AMT_ANNUITY']].sort_values(['AMT_INCOME_TOTAL','AMT_ANNUITY'], ascending=[False, True]).head(n=10)
```

#### 基礎統計運算

```python
formatStr = 'AMT_INCOME_TOTAL {:6} : {}'
AMT_INCOME_TOTAL = app_train['AMT_INCOME_TOTAL']
print(formatStr.format('max', AMT_INCOME_TOTAL.max())) 
print(formatStr.format('min', AMT_INCOME_TOTAL.min())) 
print(formatStr.format('sum', AMT_INCOME_TOTAL.sum())) 
print(formatStr.format('mean', AMT_INCOME_TOTAL.mean())) 
print(formatStr.format('median', AMT_INCOME_TOTAL.median())) 
print(formatStr.format('count', AMT_INCOME_TOTAL.count())) 
```

#### 條件查詢

```python
fliter = (AMT_INCOME_TOTAL > 9000000)
app_train[fliter].sort_values('AMT_INCOME_TOTAL', ascending=False)
```
